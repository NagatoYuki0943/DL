```python
import pickle
from pathlib import Path
from collections import Counter
```

```python
result_path = Path(
    "D:/ml/code/mm/mmdetection/work_dirs/cascade-rcnn_r50_fpn_2x_classes16/result.pkl"
)
unique_ids_path = Path(
    "D:/ml/code/mm/mmdetection/work_dirs/cascade-rcnn_r50_fpn_2x_classes16/unique_ids.pkl"
)
ids_num_path = Path(
    "D:/ml/code/mm/mmdetection/work_dirs/cascade-rcnn_r50_fpn_2x_classes16/ids_num.pkl"
)
```

```python
assert result_path.exists(), f"{result_path} not exists"
```

# load pickle

```python
with open(result_path, mode="rb") as f:
    data = pickle.load(f)
len(data)
```

```python
data[1]
```

```python
Path(data[1]["img_path"]).stem
```

```python
data[1]["pred_instances"]["labels"].tolist()
```

# get unique id list

```python
new_data = {}
for d in data:
    new_data[Path(d["img_path"]).stem] = list(
        sorted(set(d["pred_instances"]["labels"].tolist()))
    )
new_data
```

```python
with open(unique_ids_path, mode="wb") as f:
    pickle.dump(new_data, f)
```

```python
with open(unique_ids_path, mode="rb") as f:
    print(pickle.load(f))
```

# get id counter

```python
new_data = {}
for d in data:
    new_data[Path(d["img_path"]).stem] = dict(
        Counter(d["pred_instances"]["labels"].tolist())
    )
new_data
```

```python
with open(ids_num_path, mode="wb") as f:
    pickle.dump(new_data, f)
```

```python
with open(ids_num_path, mode="rb") as f:
    print(pickle.load(f))
```
