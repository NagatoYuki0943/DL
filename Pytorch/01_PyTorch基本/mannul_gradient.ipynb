```python
import numpy as np
import matplotlib.pyplot as plt
```

# $$x ^ 2$$

```python
# 目标函数
def f(x):
    return x ** 2
```

```python
# 目标函数的梯度(导数)
def g_grad(x):
    return 2 * x
```

```python
# 梯度下降
def gd(X, learning_rate):
    return X - learning_rate * g_grad(X)
```

## learning_rate = 0.1

```python
learning_rate = 0.1 # 学习率
X = 10 # X初始值
```

```python
X_list = [X]
Y_list = [f(X)]
for i in range(10):
    X = gd(X, learning_rate)
    X_list.append(X)
    Y_list.append(f(X))
    print(X)
```

```python
# 原始图像
X_ = np.linspace(-10, 10, 200).tolist()
Y_ = []
for x_ in X_:
    Y_.append(f(x_))
```

```python
plt.plot(X_, Y_)
plt.plot(X_list, Y_list)
plt.grid()
plt.show()
```

# $$x * cos(cx) \quad c为常数$$

```python
c = 0.15 * np.pi
```

```python
# 目标函数
def f(x):
    return x * np.cos(c * x)
```

```python
# 目标函数的梯度(导数)
def g_grad(x):
    return np.cos(c * x) - c * x * np.sin(c * x)
```

```python
# 损失函数
def mse(y_pred, y):
    return (y_pred - y) ** 2
```

```python
# 梯度下降
def gd(X, learning_rate):
    return X - learning_rate * g_grad(X)
```

## learning_rate = 2

```python
learning_rate = 2 # 学习率
X = 10 # X初始值
```

```python
X_list = [X]
Y_list = [f(X)]
for i in range(10):
    X = gd(X, learning_rate)
    X_list.append(X)
    Y_list.append(f(X))
    print(X)
```

```python
# 原始图像
X_ = np.linspace(-10, 10, 200).tolist()
Y_ = []
for x_ in X_:
    Y_.append(f(x_))
```

```python
plt.plot(X_, Y_)
plt.plot(X_list, Y_list)
plt.grid()
plt.show()
```
