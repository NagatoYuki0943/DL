```python
import torch
from torch import nn
import torch.nn.functional as F
```

```python
x = torch.ones(1, 1, 5, 5)
x = torch.cat([x * i for i in range(1, 4)], dim=1)
print(x.shape)
x
```

```python
# mode: [nearest, linear, bilinear, bicubic and trilinear]. Default: nearest
upsample = nn.Upsample(size=(10, 10), mode='nearest')
y1 = upsample(x)
print(y1.shape)
y1
```

```python
# mode: [nearest, linear, bilinear, bicubic and trilinear]. Default: nearest
upsample = nn.Upsample(scale_factor=2, mode='nearest')
y2 = upsample(x)
print(y2.shape)
y2
```

```python
upsample = nn.UpsamplingNearest2d(size=(10, 10))
y3 = upsample(x)
print(y3.shape)
y3
```

```python
upsample = nn.UpsamplingNearest2d(scale_factor=2)
y4 = upsample(x)
print(y4.shape)
y4
```

```python
upsample = nn.UpsamplingBilinear2d(size=(10, 10))
y5 = upsample(x)
print(y5.shape)
y5
```

```python
upsample = nn.UpsamplingBilinear2d(scale_factor=2)
y6 = upsample(x)
print(y6.shape)
y6
```

```python
# mode: [nearest, linear, bilinear, bicubic and trilinear]. Default: nearest
y7 = F.interpolate(x, size=(10, 10), mode='nearest')
print(y7.shape)
y7
```

```python
# mode: [nearest, linear, bilinear, bicubic and trilinear]. Default: nearest
y8 = F.interpolate(x, scale_factor=2, mode='nearest')
print(y8.shape)
y8
```
