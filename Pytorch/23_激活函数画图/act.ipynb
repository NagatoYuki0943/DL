```python
import torch
from torch import nn
from matplotlib import pyplot as plt
from matplotlib.pyplot import MultipleLocator
import numpy as np
```

```python
x = torch.arange(-10, 10.1, 0.1)
x
```

```python
def draw(x: np.ndarray, y: np.ndarray, title):
    plt.figure(figsize=(20, 10), dpi=100)
    plt.plot(x, y)

    # 设置刻度间隔
    x_major_locator = MultipleLocator(1)
    y_major_locator = MultipleLocator(1)
    ax = plt.gca()  # ax为两条坐标轴的实例
    # 设置xy刻度
    ax.xaxis.set_major_locator(x_major_locator)
    ax.yaxis.set_major_locator(y_major_locator)

    plt.grid()
    plt.title(title)
    plt.savefig(title + ".png")
    plt.show()
    plt.close()
```

# ReLU max(0, x)

```python
relu = nn.ReLU()
```

```python
draw(x.numpy(), relu(x).numpy(), "ReLU")
```

# ReLU6 min(max(0, x), 6)

```python
relu6 = nn.ReLU6()
```

```python
draw(x.numpy(), relu6(x).numpy(), "ReLU6")
```

# LeakyReLU

```python
leaky_relu = nn.LeakyReLU()
```

```python
draw(x.numpy(), leaky_relu(x).numpy(), "LeakyReLU")
```

# Mish

$$
\text{Mish}(x) = x * \text{Tanh}(\text{Softplus}(x))
$$

```python
mish = nn.Mish()
```

```python
draw(x.numpy(), mish(x).numpy(), "Mish")
```

# SiLU/Swish x * sigmoid(x)

$$
\text{silu}(x) = x * \sigma(x), \text{where } \sigma(x) \text{ is the logistic sigmoid.}
$$

```python
silu = nn.SiLU()
```

```python
draw(x.numpy(), silu(x).numpy(), "SiLU")
```

# Hardswish

0                   x < -3

x * (x + 3) / 6     -3 <= x <= 3

x                   x > 3

```python
hard_swish = nn.Hardswish()
```

```python
draw(x.numpy(), hard_swish(x).numpy(), "Hardswish")
```

# GELU

$$
\text{GELU}(x) = x * \Phi(x)
$$

```python
gelu = nn.GELU()
```

```python
draw(x.numpy(), gelu(x).numpy(), "GELU")
```

# Sigmoid 1 / (1 + e^(-x))

$$
\text{Sigmoid}(x) = \sigma(x) = \frac{1}{1 + \exp(-x)}
$$

```python
sigmoid = nn.Sigmoid()
```

```python
draw(x.numpy(), sigmoid(x).numpy(), "Sigmoid")
```

# Hardsigmoid

0               x < -2.5

0.2 * x + 0.5   -2.5 <= x <= 2.5

1               x > 2.5

```python
hard_sigmoid = nn.Hardsigmoid()
```

```python
draw(x.numpy(), hard_sigmoid(x).numpy(), "Hardsigmoid")
```

# Tanh

$$
\text{Tanh}(x) = \tanh(x) = \frac{\exp(x) - \exp(-x)} {\exp(x) + \exp(-x)}
$$

```python
tanh = nn.Tanh()
```

```python
draw(x.numpy(), tanh(x).numpy(), "Tanh")
```

# Softplus log(1 + e^x)



```python
softplus = nn.Softplus()
```

```python
draw(x.numpy(), softplus(x).numpy(), "Softplus")
```

# Compare

```python
def compare():
    x = torch.arange(start=-10, end=10, step=0.01)

    _relu = nn.ReLU()
    _relu6 = nn.ReLU6()
    _leaky_relu = nn.LeakyReLU()
    _sigmoid = nn.Sigmoid()
    _hard_sigmoid = nn.Hardsigmoid()
    _tanh = nn.Tanh()
    _silu = nn.SiLU()
    _hard_swish = nn.Hardswish()
    _mish = nn.Mish()
    _gelu = nn.GELU()
    _softplus = nn.Softplus()

    _relu_y = _relu(x)
    _relu6_y = _relu6(x)
    _leaky_relu_y = _leaky_relu(x)
    _sigmoid_y = _sigmoid(x)
    _hard_sigmoid_y = _hard_sigmoid(x)
    _tanh_y = _tanh(x)
    _silu_y = _silu(x)
    _hard_swish_y = _hard_swish(x)
    _mish_y = _mish(x)
    _gelu_y = _gelu(x)
    _softplus_y = _softplus(x)

    plt.figure(figsize=(30, 15), dpi=100)
    plt.plot(x.numpy(), _relu_y.numpy(), label="ReLU")
    plt.plot(x.numpy(), _relu6_y.numpy(), label="ReLU6")
    plt.plot(x.numpy(), _leaky_relu_y.numpy(), label="LeakyReLU")
    plt.plot(x.numpy(), _sigmoid_y.numpy(), label="Sigmoid")
    plt.plot(x.numpy(), _hard_sigmoid_y.numpy(), label="Hardsigmoid")
    plt.plot(x.numpy(), _tanh_y.numpy(), label="Tanh")
    plt.plot(x.numpy(), _silu_y.numpy(), label="SiLU")
    plt.plot(x.numpy(), _hard_swish_y.numpy(), label="Hardswish")
    plt.plot(x.numpy(), _mish_y.numpy(), label="Mish")
    plt.plot(x.numpy(), _gelu_y.numpy(), label="GELU")
    plt.plot(x.numpy(), _softplus_y.numpy(), label="Softplus")

    # 设置刻度间隔
    x_major_locator = MultipleLocator(1)
    y_major_locator = MultipleLocator(1)
    ax = plt.gca()  # ax为两条坐标轴的实例
    # 设置xy刻度
    ax.xaxis.set_major_locator(x_major_locator)
    ax.yaxis.set_major_locator(y_major_locator)

    plt.legend(loc="best")  # 显示图例
    plt.grid()
    plt.title("compare")
    plt.savefig("compare.png")
    plt.show()
    plt.close()
```

```python
compare()
```
