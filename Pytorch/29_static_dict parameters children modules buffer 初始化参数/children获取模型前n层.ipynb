```python
import torch
from torchvision import models
```

# 获取预训练模型

```python
x = torch.randn(1, 3, 224, 224)
```

```python
model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
model
```

# model.children() 返回的children放入Sequential中可以执行

```python
model.children()
```

```python
len(list(model.children()))
```

```python
list(model.children())
```

```python
net_childrens = torch.nn.Sequential(*list(model.children())[:-2])
net_childrens.eval()
```

```python
net_childrens(x)
```

```python
net_childrens(x).size()
```

# model.modules() 返回的module放入Sequential中不能执行,因为递归返回的

```python
model.modules()
```

```python
len(list(model.modules()))
```

```python
list(model.modules())
```

```python
net_modules = torch.nn.Sequential(*list(model.modules())[:-2])
net_modules.eval()
```

```python
net_modules(x)
```
